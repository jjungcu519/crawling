```python
log22[['keywords', 'result']]= log20['Analysis'].str.split('결과',expand=True)
log22
```

```python
# 데이터를 키워드와 결과로 나누어서 새로운 리스트로 작성합니다.
split_data = []
for item in log22:
    keywords, result = item[1].rsplit('결과:', 1)
    split_data.append([keywords.strip(), result.strip()])

# 데이터프레임으로 변환합니다.
df = pd.DataFrame(split_data, columns=['키워드', '결과'])

# CSV 파일로 저장합니다.
df.to_csv('electric_vehicle_analysis.csv', index=False, encoding='utf-8-sig')
```
